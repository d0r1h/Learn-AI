#### Transpose of a 2D Matrix

The transpose of a matrix is formed by turning its rows into columns and columns into rows.

```python
Matirx = [[1, 2, 3], [4, 5, 6]]

print(Matirx)
print(len(Matirx))
print(len(Matirx[0]))
```

__Idea__ : Currently we have a Matrix of 2 rows and 3 colmns, shape of (2,3) and when we do the transpose of this matrix, 
we need to convert this into (3,2), so we need 3 rows and 2 columns.

```python
# This is what we need as the result of any input matrix 

for i in range(len(Matirx[0])):
    print(len(Matirx) * [0])     
```

```python
# traversing though Matix 

for row in range(len(Matirx)):
    for val in range(len(Matirx[0])):
        print(Matirx[row][val])
```

```python
rows = len(Matirx)
columns = len(Matirx[0])

print(rows, columns)
```

```python
# creating an empty transposed matrix of any input matrix 

result = [rows * [0] for _ in range(columns)] 
print(result)
```

```python
# We have to take element of the row i, and column j & place them into row j, and column i  

for i in range(rows):
    for j in range(columns):
        result[j][i] = Matirx[i][j]
```

```python
# Let's understand this 

# This is what we'll get in the above loop 
print(Matirx[0][0])
print(Matirx[0][1])
print(Matirx[0][2])
print(Matirx[1][0])
print(Matirx[1][1])
print(Matirx[1][2])
print('\n')

# and if we switch the place of index then we'll get the following (aks tranpsoe)

print(Matirx[0][0])
print(Matirx[1][0])
print(Matirx[0][1])
print(Matirx[1][1])

# we can't print element (3,6) in above manner becuase that will be out of index for the Matrix
```

```python
def transpose_matrix(a: list[list[int|float]]) -> list[list[int|float]]:
    
    rows = len(a)
    columns = len(a[0])

    transpose_matrix =  [rows * [0] for _ in range(columns)]

    for i in range(rows):
        for j in range(columns):
            transpose_matrix[j][i] = a[i][j]

    return transpose_matrix
```
