```python
import numpy as np
```

```python
array = np.array([1, 2, 3, 4, 5])
array
```

__We have to implement follwoing formula for getting normalized values__

$x_{\text{norm}} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}$

```python
min_arr = min(array)
max_arr = max(array)

scaled = []

for value in array:
    scaled.append((value - min_arr)/(max_arr - min_arr))

print(f'Original array : {array}')
print(f'Normaized array : {scaled}')
```

```python
array = np.array([5, 5, 5, 5])
array

min_arr = min(array)
max_arr = max(array)

scaled = []

if len(set(array)) > 1:
    
    for value in array:
        scaled.append((value - min_arr)/(max_arr - min_arr))
else:
    scaled = [0.0] * len(array)

print(f'Original array : {array}')
print(f'Normaized array : {scaled}')
```
